```python
# ============================================================
# FFT -> EQ Gains -> IFFT -> SD test with plots
# ============================================================
%matplotlib inline
from pynq import Overlay, allocate
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
import time

BITFILE = "design_1k.bit"   # change if needed
N = 1024

ol = Overlay(BITFILE)

print("IPs:", ol.ip_dict.keys())

dma = ol.axi_dma_0
dma_cfg_fwd = ol.axi_dma_config
dma_cfg_inv = ol.axi_dma_config1
gain_gpio = ol.axi_gpio_gain


# ============================================================
# Gain helpers
# ============================================================

def gain_to_q15(gain):
    """
    Convert float gain to unsigned Q1.15-style value.
    1.0 = 32768
    0.5 = 16384
    1.5 = 49152
    """
    val = int(round(gain * 32768))
    return max(0, min(65535, val))


def set_eq_gains(low=1.0, mid=1.0, high=1.0):
    low_q  = gain_to_q15(low)
    mid_q  = gain_to_q15(mid)
    high_q = gain_to_q15(high)

    # Channel 1:
    # bits [15:0]  = low
    # bits [31:16] = mid
    gain_gpio.channel1.write((mid_q << 16) | low_q, 0xFFFFFFFF)

    # Channel 2:
    # bits [15:0] = high
    gain_gpio.channel2.write(high_q, 0x0000FFFF)

    print(f"Set gains: low={low} ({low_q}), mid={mid} ({mid_q}), high={high} ({high_q})")


# ============================================================
# Packing helpers
# ============================================================

def pack_complex_i16(real, imag=None):
    real = np.asarray(real, dtype=np.int16)

    if imag is None:
        imag = np.zeros_like(real, dtype=np.int16)
    else:
        imag = np.asarray(imag, dtype=np.int16)

    return (
        real.astype(np.uint16).astype(np.uint32)
        | (imag.astype(np.uint16).astype(np.uint32) << 16)
    )


def unpack_complex_i16(raw):
    raw = np.asarray(raw, dtype=np.uint32)

    real = (raw & 0xFFFF).astype(np.uint16).view(np.int16)
    imag = ((raw >> 16) & 0xFFFF).astype(np.uint16).view(np.int16)

    return real, imag


# ============================================================
# Main hardware run
# ============================================================

def run_fft_eq_ifft(bin_num, amp=8):
    """
    Runs one input sine through:
    DMA -> FFT -> EQ -> IFFT -> SD -> DMA

    Returns:
        real_in, real_out, imag_out, scale, corr
    """

    cfg_fwd = allocate(shape=(1,), dtype=np.uint16)
    cfg_inv = allocate(shape=(1,), dtype=np.uint16)
    inp = allocate(shape=(N,), dtype=np.uint32)
    out = allocate(shape=(N,), dtype=np.uint32)

    # Forward and inverse FFT config
    cfg_fwd[0] = 0x0001
    cfg_inv[0] = 0x0000

    t = np.arange(N)
    real_in = (amp * np.sin(2 * np.pi * bin_num * t / N)).astype(np.int16)

    inp[:] = pack_complex_i16(real_in)
    out[:] = 0

    cfg_fwd.flush()
    cfg_inv.flush()
    inp.flush()
    out.flush()

    # Start receive first
    dma.recvchannel.transfer(out)

    # Send FFT configs
    dma_cfg_fwd.sendchannel.transfer(cfg_fwd)
    dma_cfg_inv.sendchannel.transfer(cfg_inv)

    dma_cfg_fwd.sendchannel.wait()
    dma_cfg_inv.sendchannel.wait()

    # Send data
    dma.sendchannel.transfer(inp)

    dma.sendchannel.wait()
    dma.recvchannel.wait()

    out.invalidate()

    real_out, imag_out = unpack_complex_i16(out)

    x = real_in.astype(np.float64)
    y = real_out.astype(np.float64)

    scale = np.dot(x, y) / max(np.dot(x, x), 1)
    corr = np.corrcoef(x, y)[0, 1] if np.std(x) > 0 and np.std(y) > 0 else None

    return real_in, real_out, imag_out, scale, corr


# ============================================================
# Plot helper
# ============================================================

def plot_fft_eq_result(label, bin_num, real_in, real_out, imag_out, scale, corr, save=False):
    x = real_in.astype(np.float64)
    y = real_out.astype(np.float64)

    # Output spectrum after IFFT result
    out_fft = np.fft.fft(y)
    out_mag = np.abs(out_fft)

    # Input spectrum reference
    in_fft = np.fft.fft(x)
    in_mag = np.abs(in_fft)

    fig = plt.figure(figsize=(14, 12))

    # --------------------------------------------------------
    # 1. Input waveform
    # --------------------------------------------------------
    plt.subplot(4, 1, 1)
    plt.plot(real_in[:256], marker="o", markersize=2)
    plt.title(f"{label} — Input Waveform, Bin {bin_num}")
    plt.xlabel("Sample")
    plt.ylabel("Amplitude")
    plt.grid(True)

    # --------------------------------------------------------
    # 2. Output waveform
    # --------------------------------------------------------
    plt.subplot(4, 1, 2)
    plt.plot(real_out[:256], marker="o", markersize=2)
    plt.title(f"{label} — Output Waveform after FFT → EQ → IFFT → SD")
    plt.xlabel("Sample")
    plt.ylabel("Amplitude")
    plt.grid(True)

    # --------------------------------------------------------
    # 3. Input vs output overlay
    # --------------------------------------------------------
    plt.subplot(4, 1, 3)

    # Scale input by measured output/input scale for comparison
    plt.plot(y[:256], label="Hardware Output", linewidth=2)
    plt.plot((x * scale)[:256], "--", label=f"Input × measured scale ({scale:.2f})")

    plt.title(f"{label} — Input vs Output Overlay")
    plt.xlabel("Sample")
    plt.ylabel("Amplitude")
    plt.legend()
    plt.grid(True)

    # --------------------------------------------------------
    # 4. Output FFT magnitude
    # --------------------------------------------------------
    plt.subplot(4, 1, 4)
    plt.plot(out_mag, label="FFT of Hardware Output")
    plt.plot(in_mag * abs(scale), "--", label="Input FFT × measured scale", alpha=0.7)

    plt.axvline(bin_num, linestyle="--", label=f"+bin {bin_num}")
    plt.axvline(N - bin_num, linestyle="--", label=f"-bin {N - bin_num}")

    plt.title(f"{label} — Output Spectrum")
    plt.xlabel("FFT Bin")
    plt.ylabel("Magnitude")
    plt.legend()
    plt.grid(True)

    plt.tight_layout()

    if save:
        filename = label.lower().replace(" ", "_").replace("/", "_") + ".png"
        plt.savefig(filename, dpi=150)
        print(f"Saved plot: {filename}")

    display(fig)
    plt.show()
    plt.close(fig)

    print("\nResult summary:")
    print(f"Label: {label}")
    print(f"Bin: {bin_num}")
    print(f"Estimated output/input scale: {scale:.3f}")
    print(f"Correlation: {corr}")
    print(f"Max abs input: {np.max(np.abs(real_in))}")
    print(f"Max abs output: {np.max(np.abs(real_out))}")

    # Print strongest bins in output spectrum
    top_bins = np.argsort(out_mag)[-8:][::-1]
    print("Top output FFT bins:")
    for b in top_bins:
        print(f"  bin {b:4d}: mag={out_mag[b]:10.3f}")


# ============================================================
# Combined test helper
# ============================================================

def run_and_plot(label, bin_num, amp=8, save=False):
    print("\n" + "=" * 70)
    print(label)
    print(f"Bin: {bin_num}")
    print("=" * 70)

    real_in, real_out, imag_out, scale, corr = run_fft_eq_ifft(bin_num, amp=amp)

    print(f"Estimated output/input scale: {scale:.3f}")
    print(f"Correlation: {corr}")
    print(f"Max abs input: {np.max(np.abs(real_in))}")
    print(f"Max abs output: {np.max(np.abs(real_out))}")

    print("Input first 16:")
    print(real_in[:16])

    print("Output real first 16:")
    print(real_out[:16])

    print("Output imag first 16:")
    print(imag_out[:16])

    plot_fft_eq_result(
        label=label,
        bin_num=bin_num,
        real_in=real_in,
        real_out=real_out,
        imag_out=imag_out,
        scale=scale,
        corr=corr,
        save=save
    )

    return scale


# ============================================================
# Run tests
# ============================================================

# -----------------------------
# Unity baseline
# -----------------------------
set_eq_gains(low=1.0, mid=1.0, high=1.0)

s_low_base = run_and_plot(
    label="LOW baseline unity gain",
    bin_num=3,
    amp=8,
    save=False
)

s_mid_base = run_and_plot(
    label="MID baseline unity gain",
    bin_num=50,
    amp=8,
    save=False
)

s_high_base = run_and_plot(
    label="HIGH baseline unity gain",
    bin_num=200,
    amp=8,
    save=False
)


# -----------------------------
# Low boost
# -----------------------------
set_eq_gains(low=1.5, mid=1.0, high=1.0)

s_low_boost = run_and_plot(
    label="LOW boosted 1.5x",
    bin_num=3,
    amp=8,
    save=False
)

s_mid_after_low_boost = run_and_plot(
    label="MID after low boost",
    bin_num=50,
    amp=8,
    save=False
)

s_high_after_low_boost = run_and_plot(
    label="HIGH after low boost",
    bin_num=200,
    amp=8,
    save=False
)


# -----------------------------
# Mid cut
# -----------------------------
set_eq_gains(low=1.0, mid=0.5, high=1.0)

s_low_after_mid_cut = run_and_plot(
    label="LOW after mid cut",
    bin_num=3,
    amp=8,
    save=False
)

s_mid_cut = run_and_plot(
    label="MID cut 0.5x",
    bin_num=50,
    amp=8,
    save=False
)

s_high_after_mid_cut = run_and_plot(
    label="HIGH after mid cut",
    bin_num=200,
    amp=8,
    save=False
)


# ============================================================
# Final summary
# ============================================================

print("\n" + "=" * 70)
print("FINAL SUMMARY")
print("=" * 70)

print(f"LOW baseline:        {s_low_base:.3f}")
print(f"LOW boosted 1.5x:    {s_low_boost:.3f}")
print(f"LOW boost ratio:     {s_low_boost / s_low_base:.3f}")

print()

print(f"MID baseline:        {s_mid_base:.3f}")
print(f"MID cut 0.5x:        {s_mid_cut:.3f}")
print(f"MID cut ratio:       {s_mid_cut / s_mid_base:.3f}")

print()

print(f"HIGH baseline:       {s_high_base:.3f}")
print(f"HIGH after tests:    {s_high_after_mid_cut:.3f}")
print(f"HIGH ratio:          {s_high_after_mid_cut / s_high_base:.3f}")
```
